```python
# Берем блоки из листа срезами и даем им соответсвующие названия
debit_block_train = df_raw.iloc[2:15, 1:] # Дебит
pressure_block_train = df_raw.iloc[17:30, 1:] # Давления
zpa_pressure_block_train = df_raw.iloc[32:35, 1:] # ЗПА давление
temp_block_train = df_raw.iloc[37:50, 1:] # Температура
zpa_temp_block_train = df_raw.iloc[52:55, 1:] # ЗПА температура
```

```python
# Берем строку с датами
dates_train = df_raw.iloc[0, 1:]

# Берем названия кустов
kust_names_train = df_raw.iloc[2:15, 0].values
```

```python
# Создаем функцию для того, чтобы перевести данные в нужный нам формат
def to_right_format(df_block, parameter_name):
    df = pd.DataFrame(df_block.values, columns=dates_train)
    df["Куст"] = kust_names_train # создаем колонку "Куст" с названиями кустов
    df_long = df.melt(id_vars=["Куст"], var_name="Дата", value_name=parameter_name)
    return df_long
```

```python
# Используем функцию на блоках где 13 строк
df_debit_train = to_right_format(debit_block_train, "Дебит")
df_pressure_train = to_right_format(pressure_block_train, "Давление")
df_temp_train = to_right_format(temp_block_train, "Температура")
```

```python
# Функция для того, чтобы сделать колонку с соответсвующим кустам коллекторам
def assign_collector(kust_name):
    if kust_name in ["куст 1", "куст 2", "куст 3", "куст 4", "куст 5", "куст 6"]:
        return "коллектор №1"
    elif kust_name in ["куст 7", "куст 8", "куст 9", "куст 10"]:
        return "коллектор №2"
    else:
        return "коллектор №3"
```

```python
def merge_zpa_features(main_df, zpa_pressure_long, zpa_temp_long):
    # Присоединяем давление на входе в ЗПА
    df = main_df.merge(
        zpa_pressure_long,
        how="left",
        left_on=["Коллектор", "time_index"],
        right_on=["collector", "time_index"] # Указываем по каким колонкам делаем объединение
    ).drop(columns=["collector"])  # Удаляем дублирующий столбец

    # Присоединяем температуру на входе в ЗПА
    df = df.merge(
        zpa_temp_long,
        how="left",
        left_on=["Коллектор", "time_index"],
        right_on=["collector", "time_index"]
    ).drop(columns=["collector"])

    return df
```

```python
# Функция для замены выбросов(в том числе и тех NaN, что мы заменили на нули) на медиану предыдщих значений по каждому кусту
def replace_outliers_with_previous_median(df, column, condition):
    df_copy = df.copy() # делаем копию, чтобы не изменять оригинальный датасет

    for well in df_copy["Куст"].unique(): # для каждого куста в списке уникальных значений кустов...
        well_df = df_copy[df_copy["Куст"] == well] # создаем отдельный датасет...
        indices = well_df[condition(well_df[column])].index # и берем индексы значений, которые соотвествуют условию(condition) в колонке(column) датасета

        for idx in indices:
            previous_values = well_df.loc[:idx - 1, column] # берем все предыдущие значения до текущей даты для каждого индекса...
            previous_valid = previous_values[~condition(previous_values)] # которые не являются выбросами

            if not previous_valid.empty: # и если есть такие значения...
                median_val = previous_valid.mean() # считаем медиану...
                df_copy.loc[idx, column] = median_val # и заменяем выброс на медиану

    return df_copy
```

```python
def classify_hydrate_risk(df, A=4, B=2750, C=46.5):
    """
    Классифицирует каждый объект датасета по риску гидратообразования метана.
    
    Параметры:
    - df: DataFrame, обязательно с колонками 'Температура' (°C) и 'Давление' (атм или аналогично)
    - A, B, C: параметры уравнения гидратообразования (можно подставить свои)
    
    Возвращает:
    - DataFrame с добавленной колонкой 'Риск гидратообразования'
    """
    
    # Перевод температуры в Кельвины
    temp_K = df['Температура'] + 273.15
    
    # Вычисляем давление кривой гидратообразования
    lnP = A - B / (temp_K - C)
    P = np.exp(lnP)
    
    # Верхняя и нижняя границы
    P_up = P * 1.15
    P_lo = P * 0.85
    
    # Пустой список для статусов
    risk_status = []
    
    for i in range(len(df)):
        pressure = df.iloc[i]['Давление']
        
        if pressure > P_up.iloc[i]:
            risk_status.append('Гидратообразование')
        elif pressure < P_lo.iloc[i]:
            risk_status.append('Нет риска')
        else:
            risk_status.append('Потенциальный риск')
    
    # Добавляем новую колонку
    df['Риск гидратообразования'] = risk_status
    
    return df
```
